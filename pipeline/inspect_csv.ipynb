```python
# Try with semicolon delimiter first
try:
    df_semicolon = pd.read_csv("ETH_day.csv", delimiter=';')
    print("CSV with semicolon delimiter:")
    print(df_semicolon.head())
    print("\nColumns:", df_semicolon.columns.tolist())
    print("\nShape:", df_semicolon.shape)
except Exception as e:
    print(f"Error with semicolon delimiter: {e}")

# Try with comma delimiter
try:
    df_comma = pd.read_csv("ETH_day.csv", delimiter=',')
    print("\n\nCSV with comma delimiter:")
    print(df_comma.head())
    print("\nColumns:", df_comma.columns.tolist())
    print("\nShape:", df_comma.shape)
except Exception as e:
    print(f"Error with comma delimiter: {e}")

# Try with auto-detection
try:
    df_auto = pd.read_csv("ETH_day.csv", sep=None, engine='python')  # Auto-detect delimiter
    print("\n\nCSV with auto-detected delimiter:")
    print(df_auto.head())
    print("\nColumns:", df_auto.columns.tolist())
    print("\nShape:", df_auto.shape)
except Exception as e:
    print(f"Error with auto-detection: {e}")
```

```python
# Once we have a properly parsed dataframe
try:
    print("\n\nDataframe info:")
    print(df.info())

    print("\n\nMissing values:")
    print(df.isnull().sum())

    # Identify numeric columns
    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    print("\n\nNumeric columns:", numeric_cols)

    # Show some basic statistics
    print("\n\nBasic statistics:")
    print(df[numeric_cols].describe())

    # Plot a sample of data if price columns are found
    price_candidates = [col for col in df.columns if any(term in col.lower() for term in ['close', 'price', 'open', 'high', 'low'])]
    if price_candidates:
        plt.figure(figsize=(12, 6))
        df[price_candidates[0]].plot(title=f"{price_candidates[0]} Values")
        plt.show()
except NameError:
    print("Could not create a proper DataFrame for analysis.")
```
